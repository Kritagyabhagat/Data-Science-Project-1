##### `1. Dataset Selection:`


* **Dataset:** I use the "Student_report" dataset. It contains information about the Roll_number, Student_name, Subject_name, and Student_performances

##### `2. Data Loading:`

```python
import pandas as pd
df = pd.read_csv("Student_report.csv") # Load the Student_report dataset from a CSV file
df
```

##### `3. Data Cleaning:`

* Check for Missing Values:

```python
missing_values = df.isnull().sum() # Check for missing values in each column
missing_values
```

* Handle Missing Values:

```python
a = df.fillna(0) # Fill missing values with 0
a
```

```python
b = df.dropna() # Drop rows with missing values
b
```

* Check for Duplicates:

```python
duplicated_value = df.duplicated() # Check for duplicates in each row
duplicated_value
```

* Remove Duplicates:

```python
df.drop_duplicates(inplace=True) # Remove duplicates if any
```

##### `4. Data Manipulation:`

* Create New Columns:

```python
df['Percentage'] = (df['Total']/900) * 100 # New column value
print(df['Percentage'])
```

* Filter Data:

```python
df_pass = df[df['Student_performances'] == 'Pass']
print(df_pass)
```

```python
df_fail = df[df['Student_performances'] == 'Fail']
print(df_fail)
```

* Group Data:

```python
grouped_by_Student_performances = df.groupby('Student_performances')
print(grouped_by_Student_performances)
```

##### `5. Data Analysis:`

* Summary Statistics:

```python
statistics = df.describe()
print(statistics)
```

* Correlation Analysis:

```python
# correlation = df.corr()
```

* Visualization (Optional):

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.pairplot(df, hue='Student_performances')
plt.show()
```
