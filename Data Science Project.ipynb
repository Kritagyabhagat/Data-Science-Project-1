## <div style="padding: 10px 10px; color: blue; border-left:3px solid red">2. Data Loading:</div>

## <div style="padding: 10px 10px; color: blue; border-left:3px solid red">5. Data Analysis:</div>

* ### Library

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

* ### Summary Statistics:

```python
statistics = df.describe()
print(statistics)
```

* ### BAR graph

```python
plt.figure(figsize=(19000, 1200))
df.plot.barh('Name','Percentage', color="b", fontsize = 5)  # Horizontal bars with black color

# Add grid lines on the y-axis with red color
plt.grid(axis="x", color="r")

# Rotate x-axis labels by 90 degrees for better readability
plt.xticks(rotation=90)

# Title and labels with clear formatting
plt.title('\n Percentage of Students\n', fontsize=40)
plt.ylabel("Student Name", fontsize=30)
plt.xlabel("Percentage", fontsize=30)

# Display the plot
plt.show()
```

* ### PIE graph

```python
df_filtered = df.dropna(subset=['Performance'])  # Remove rows with missing values in Performance
chart = df_filtered['Performance'].value_counts().sort_values(ascending=False)  # Count occurrences
plt.pie(chart, labels=chart.index.to_list(), autopct="%1.1f%%")  # Add percentages to labels
plt.title("Showing Performance Statistics of Student", color='red')
plt.show()
```

* ### Plotting

```python
sns.pairplot(df)
plt.show()
```
